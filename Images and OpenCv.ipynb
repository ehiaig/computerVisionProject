```python
plt.imshow(img)
```

As you see, the img is rendered in a blue tone. This is because opencv  and matplotlib both expect different orders of the RBG channels. 
#### matplotlib order is Red Blue Green
#### OpenCv order is Blue Green Red 

So the image above, opencv rendered it in BGR format so we need to render it back in OpenCv format using cvtColor() function:

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

Now we have the real image.
say we want to show it in grey scale

```python
img_gray = cv2.imread('Photos/messi.jpeg', cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
img_gray.shape
```

As you see, the shape is no longer in a 3-color channel

```python
print(img_gray.min())
print(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

#### Resizing Images

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
rez_img = cv2.resize(fix_img, (1000, 400))
plt.imshow(rez_img)
```

```python
#Say we need to resize by ratio (1.e making it 30% smaller)
w_ratio = 0.3
h_ratio = 0.3

new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
```

### Flip Image along the horizontal or vertical axis

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 1)
plt.imshow(new_img)
```
