```python
mat = np.arange(0,100).reshape(10,10)
mat
```

```python
 #let's get all values in col 1
    
mat[:, 1]
```

```python
#we can reshape it to display in col format
mat[:, 1].reshape(10, 1)
```

```python
mat[0:3, 0:5]
```

```python
#we can reassign the valus in a matrix
mat[0:3, 0:5] = 0
mat
```

```python
#we can also make a copy of an array and reassign it
newmat = mat.copy()
newmat[0:6, :] = 200
newmat
```

# Numpy and Images 

Every image has 3 channels: Red, Blue, Green

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('../Photos/messi.jpeg')
pic
```

```python
type(pic)
```

```python
#Let's transaform the photo into an array since that's eat the computer understand
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
#let's access only the red channel
pic_red = pic_arr.copy()
plt.imshow(pic_red)

#R G B channel
#we will zero out the blue channels to be left with onlu RED
plt.imshow(pic_red[:, :, 0])
```

```python
#This is the RED CHANNEL VALUES 0-255
#0 means no red, pure black. 255 means full red no black
plt.imshow(pic_red[:, :, 0], cmap='gray')
```

```python
#This is the GREEN CHANNEL VALUES 0-255
#0 means no green, pure black. 255 means full green no black

plt.imshow(pic_red[:, :, 1], cmap='gray')
```

```python
#This is the BLUE CHANNEL VALUES 0-255
plt.imshow(pic_red[:, :, 2], cmap='gray')
```

```python
#Let's do a little experiemt to remove the green
pic_red[:,:,1] = 0
plt.imshow(pic_red)
```

We can see the image begins to tend towards purple. And which is a mixure of red and blue

```python
pic_red[:,:,2] = 0
plt.imshow(pic_red)
```
