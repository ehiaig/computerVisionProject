### Blending and Pasting Images
Blending image is done through addWeighted function that uses both images and combines them,
To blend images, we use a simple formula: 
- new_pixel = alpha X pixel_1 + beta X pixel_2 + gamma

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('Photos/1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Photos/water_coins.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

##### Let's begin by resizing them to be same size

```python
img1 = cv2.resize(img1, (400,400))
plt.imshow(img1)
```

```python
img2 = cv2.resize(img2, (400,400))
plt.imshow(img2)
```

Now that both images are same size, we can then blend them together usung the addWeighted() function.

addWeighted function only works if the images are same size

```python
blended = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.imshow(blended)
```

We can increase and decrease the alpha, beta values as you want

### Overlaying images of different sizes using Numpy reassignment

```python
img1 = cv2.imread('Photos/1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Photos/water_coins.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
#renaming the variables
small_img = img1
large_img = img2

# Using Numpy slicing
x_offset = 0
y_offset = 0

x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img
plt.imshow(large_img)
```

### Blending together images of different sizes

```python
#Reloading the new images

img1 = cv2.imread('Photos/dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Photos/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
#Let's resize img2 to make it smaller than img1 since it's img2 we want to blend inside img1
img2 = cv2.resize(img2, (600,600))
```

```python
plt.imshow(img2)
```

Now let's decide where in img1 we want to blend img2. 

We do this by figuring out a 600 by 600 space inside img1 where we want the blend to occur. This is the region of interest(roi)

```python
img1.shape
```

```python
img2.shape
```

```python
x_offset = 934 - 600 #we are subtracting x axis of the smaller image from the bigger on
y_offset = 1401 - 600
```

```python
rows, cols, channels = img2.shape
```

```python
rows
```

```python
cols
```

```python
roi = img1[y_offset:1401, x_offset:943]
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
mask_inv.shape
```

Notice that mask_inv.shape is no longer a 3-channel. Let's fix it using numpy

```python
import numpy as np
```

```python
white_background = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
white_background.shape
```

```python
bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg)
```

Now let's get the roi and blend it wih the mask

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]]= small_img
```

```python
plt.imshow(large_img)
```
